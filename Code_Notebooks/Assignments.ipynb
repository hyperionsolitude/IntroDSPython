```python
def perfectNumber(x):
    
    Divisors = []
    for i in range(1, x):
        if x%i == 0:
            Divisors.append(i)    
    #print(Divisors, sum(Divisors))
    if (sum(Divisors)== x):
        return True
    else:
        return False
```

```python
def findPermuation(List):
    
    new = []
    if len(List) == 1:
        return [List]
    else:
        Element = List[0]
        tempList2 = List[1:] 
        permutations = findPermuation(tempList2)
        for i in permutations:
            for j in range(len(i)):
                temp = i.copy()
                temp.insert(j,Element)
                if(temp) not in new:
                    new.append(temp)
            temp = i.copy()
            temp.append(Element)
            if(temp) not in new:
                new.append(temp)
        return new
```

```python
def knapSack(Volume, Weight, Value, n):
  
    if n == 0 or Volume == 0:
        return 0
    if (Weight[n-1] > Volume):
        return knapSack(Volume, Weight, Value, n-1)
    else:
        return max(Value[n-1] + knapSack(Volume-Weight[n-1], Weight, Value, n-1), knapSack(Volume, Weight, Value, n-1))
  
```
