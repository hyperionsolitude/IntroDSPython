```python
# Simple is_even function definition
def is_even( i ):
    """ 
    Input: i, a positive int
    Returns True if i is even, otherwise False
    """
    remainder = i % 2
    return remainder == 0

```

```python
#########################
## EXAMPLE: applying functions to repeat same task many times
#########################
def bisection_cuberoot_approx(x, epsilon):
    """
    Input: x, an integer
    Uses bisection to approximate the cube root of x to within epsilon
    Returns: a float approximating the cube root of x
    """

    low = 0.0
    high = x
    guess = (high + low)/2.0
    while abs(guess**3 - x) >= epsilon:
        if guess**3 < x:
            low = guess
        else:
            high = guess
        guess = (high + low)/2.0
    return guess


    
```

```python
cuberoot = bisection_cuberoot_approx(1000, 0.000000000001)
```

```python
cuberoot
```

```python
x = 1
while x <= 10000:
    print(x)
    guess = bisection_cuberoot_approx(x, 0.01)
    print(guess, "is close to cube root of", x)
    x += 1
```

```python
import math
```

```python
def findDivis(x, i):
     
    return x % i == 0
```

```python
def isPerfectNumber(x):

    summation = 0
    List = []
    for i in range(1, x):
        if findDivis(x, i):
            summation += i 
            List.append(i)
    return x == summation, List
```

```python
def isPerfectNumber2(x):

    summation = 1
    List = [1]
    sqroot = math.sqrt(x)
    sqroot = int(sqroot)
    for i in range(2, sqroot+1):
        if findDivis(x, i):
            summation += i 
            List.append(i)
            a = x/i
            summation += a
            List.append(int(a))
            
    return x == summation, List
```

```python
def isPerfectNumber3(x):

    summation = 1
    List = [1]
    sqroot = math.sqrt(x)
    sqroot = int(sqroot)
    for i in range(2, sqroot+1):
        if findDivis(x, i):
            summation += i 
            List.append(i)
            a = x/i
            summation += a
            List.append(int(a))
        if summation>x:
            break
            
    return x == summation, List
```

```python
isPerfectNumber2(6)
```

```python
for i in range(2, 100001):
    flag, List = isPerfectNumber3(i)
    if flag:
        print(i, List)
```
