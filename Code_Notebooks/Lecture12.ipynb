```python
def bubble_sort(L):
    swap = False
    while not swap:
        #print('bubble sort: ' + str(L))
        swap = True
        for j in range(1, len(L)):
            if L[j-1] > L[j]:
                #print("Swaping", L[j-1], L[j])       
                swap = False
                temp = L[j]
                L[j] = L[j-1]
                L[j-1] = temp
                #print(L)
```

```python
def selection_sort(L):
    suffixSt = 0
    while suffixSt != len(L):
        #print('selection sort: ' + str(L))
        for i in range(suffixSt+1, len(L)):
            #print(L[i], L[suffixSt])
            if L[i] < L[suffixSt]:
                #print("Now swapping", L[i], L[suffixSt])
                L[suffixSt], L[i] = L[i], L[suffixSt]
        suffixSt += 1

        

       
```
